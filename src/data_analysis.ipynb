# Brazilian Cities - Data Analysis

```python
# Let's load necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
# Let's load the data file
data = pd.read_csv('../data/BRAZIL_CITIES_CLEANED.csv')
```

```python
# I want to see what is the range for IDHM 
# but only those values that are greater than 0
print(f"IDHM ranges from {data.loc[data['IDHM'] > 0, 'IDHM'].min()} to {data.loc[data['IDHM'] > 0, 'IDHM'].max()}")

# How many cities have IDHM equal to 0
num_cities_idhm_zero = (data['IDHM'] == 0).sum()
print(f"Number of cities with IDHM equal to 0: {num_cities_idhm_zero}")
```

```python
# I want to see what is the range for the population
print(f"Population ranges from {data['ESTIMATED_POP'].min()} to {data['ESTIMATED_POP'].max()}")

# How many cities have population equal to 0
num_cities_population_zero = (data['ESTIMATED_POP'] == 0).sum()
print(f"Number of cities with population equal to 0: {num_cities_population_zero}")
```

```python
# I want to investigate the list of cities flagged as capital cities
# Ordered by state, show only states with more than 1 capital city flagged
capital_cities = data[data['CAPITAL'] == 1]
capital_cities_count = capital_cities['STATE'].value_counts()
states_with_multiple_capitals = capital_cities_count[capital_cities_count > 1]
print("States with more than 1 capital city flagged:")  
for state in states_with_multiple_capitals.index:
    cities_in_state = capital_cities[capital_cities['STATE'] == state]['CITY'].tolist()
    print(f"{state}: {', '.join(cities_in_state)}")
```
